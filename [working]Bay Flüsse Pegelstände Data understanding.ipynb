We start bei loading the data in a DataFrame.

We use pandas parameters *parse_dates* to directly read the datetime column as such and *na_values* to automatically replace the '---' string from the website with NaN values.

```python
#Load the complete csv to dataframe
for file in os.listdir():
    if 'bay_river_pegel_' and 'csv' in file:
        waterlevel_hist = pd.read_csv(file, parse_dates=True, na_values= '---')
        break
    
    else:
        waterlevel_hist = None
```
